ImbalancedDataset:
A dataset with un unequal distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# print the first 5 rows of the dataframe
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 --> Legit Transactions
1 --> Fraudulent transactions

```python
# separating the legit and fradulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print (legit.shape)
```

```python
print(fraud.shape)
```

Under-sampling

Builf a sample dataset containing similar distribution of Legit & Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the TWO DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
