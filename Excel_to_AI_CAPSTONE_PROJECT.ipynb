## Question: Who are the top 7 highest paid athletes of all time?

**Hint:** please refer to [Day 2's notebook](https://colab.research.google.com/drive/1juOdCiSE9YdF3_bXYYVdi-q4qde04pgR) - section: What are the 10 products with the highest sales? (use the index on the left panel to find that section). *Note that some athletes names could appear more than once in this dataset*



Please produce a table and a plot that answer this question (you can use any visualization tool - Pandas, Seaborn, Matplotlib)

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Produce a table that shows who are the highest paid athletes
atl[["Name", "earnings_Usmillion"]].groupby(by="Name").sum().reset_index().sort_values('earnings_Usmillion', ascending=False).head(7)
top_7_paid = atl[["Name", "earnings_Usmillion"]].groupby(by="Name").sum().reset_index().sort_values('earnings_Usmillion', ascending=False).head(7)
top_7_paid
```

```python
# Create a bar plot that visualizes the results
fig, ax = plt.subplots(figsize=(10, 7))

ax = sns.barplot(x="Name", y="earnings_Usmillion", data = top_7_paid)
set_ticks = ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
```
